```python
import pandas as pd

orders_df = pd.read_csv("/content/orders (1).csv")
orders_df.head()

```

```python
users_df = pd.read_json("users.json")
users_df.head()
```

```python
import sqlite3

# Create SQLite DB
conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

# Read SQL file
with open("restaurants.sql", "r") as file:
    sql_script = file.read()

cursor.executescript(sql_script)

restaurants_df = pd.read_sql("SELECT * FROM restaurants", conn)
conn.close()

restaurants_df.head()
```

```python
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)

print(" Final dataset created: final_food_delivery_dataset.csv")
```

```python
import pandas as pd

df = pd.read_csv("/content/final_food_delivery_dataset.csv")
df.head()
```

```python
df[df["membership"] == "Gold"] \
  .groupby("city")["total_amount"] \
  .sum() \
  .sort_values(ascending=False) \
  .head(1)
```

```python
df.groupby("cuisine")["total_amount"] \
  .mean() \
  .sort_values(ascending=False) \
  .head(1)
```

```python
user_spend = df.groupby("user_id")["total_amount"].sum()
user_spend[user_spend > 1000].count()
```

```python
df["rating_range"] = pd.cut(
    df["rating"],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=["3.03.5", "3.64.0", "4.14.5", "4.65.0"]
)

df.groupby("rating_range", observed=False)["total_amount"] \
  .sum() \
  .sort_values(ascending=False) \
  .head(1)
```

```python
df[df["membership"] == "Gold"] \
  .groupby("city")["total_amount"] \
  .mean() \
  .sort_values(ascending=False) \
  .head(1)
```

```python
df.groupby("cuisine").agg(
    restaurant_count=("restaurant_name_y", "nunique"),
    total_revenue=("total_amount", "sum")
).sort_values(
    ["restaurant_count", "total_revenue"],
    ascending=[True, False]
).head(1)
```

```python
round(
    (df[df["membership"] == "Gold"].shape[0] / df.shape[0]) * 100
)
```

```python
df.groupby("restaurant_name_y").agg(
    avg_order_value=("total_amount", "mean"),
    total_orders=("user_id", "count")
).query("total_orders < 20") \
 .sort_values("avg_order_value", ascending=False) \
 .head(1)
```
